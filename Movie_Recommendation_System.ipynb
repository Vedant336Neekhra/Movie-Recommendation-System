```python
df_combined['cast']=df_combined['cast'].apply(literal_eval)
```

```python
# getting top 5 cast names only
def get_top_5_cast_names(cast_list):
    # ensuring the list is not empty and has at least 5 items
    if cast_list and len(cast_list) >= 5:
        return ' '.join(c['name'] for c in cast_list[:5])
    # if there are less than 5 items then returning all elements
    return ' '.join(c['name'] for c in cast_list)
```

```python
df_combined['cast']=df_combined['cast'].apply(get_top_5_cast_names)
```

```python
df_combined['crew']=df_combined['crew'].apply(literal_eval)  # converting the string to list
```

```python
# getting only director name from the crew
def get_director_name(crew):
  return ' '.join(c['name'] for c in crew if c['job']=='Director') # iterates in each dictionary and checks if job is director then returns it
  # i have used .join because earlier i was thinking of usinf producers name also but i realised that it would not affect much
```

```python
df_combined['crew']=df_combined['crew'].apply(get_director_name)
```

```python
df_combined.loc[df_combined['keywords'] == '', 'keywords'] = np.nan
df_combined.loc[df_combined['cast'] == '', 'cast'] = np.nan
df_combined.loc[df_combined['crew'] == '', 'crew'] = np.nan
# removing empty strings from each column
```

```python
df_combined=df_combined.dropna()
```

```python
# creating one string out of all the attributes to compare at once
def soup(df):
  return ''.join(df['genres'] + df['cast'] + df['keywords'] + df['crew'])
```

```python
df_combined['soup']=df_combined.apply(soup,axis=1)
```

```python
df_combined=df_combined.drop(['genres','cast','crew','keywords','id'],axis=1) # removing other columns which have no use now
```

```python
df_combined.index=range(len(df_combined))   # changing index to be in sequence only
```

```python
embeddings=model.encode(df_combined['soup'])   # creating embeddings
co_sim=util.cos_sim(embeddings,embeddings)     # calculating cosine similarity between each movie embedding
```

```python
co_sim = pd.DataFrame(co_sim,index=df_combined['original_title'],columns=df_combined['original_title'])  # a dataframe with cosine similarity matrix and index as movie title
```

Creating input interface using gradio

```python
def output(movie_name):
  movie_name = movie_name.title()
  l=(co_sim2[[movie_name]]*0.4 + co_sim[[movie_name]]*0.6).nlargest(6,columns=movie_name).index.to_list()
  return l[1],l[2],l[3],l[4],l[5]
```

```python
import gradio as gr

# creating an interface to give text input and 5 top movie names related to it
interface = gr.Interface(fn=output,inputs="text",outputs=['text','text','text','text','text'],title="Movie Recommendation System",description="Enter a movie title to get suggestions for similar movies.")

interface.launch()
```
